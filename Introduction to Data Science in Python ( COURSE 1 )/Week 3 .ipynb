```python
#cwurData.csv
df = pd.read_csv('cwurData.csv')
```

```python
df['world_rank'].unique()
```

```python
df.columns
```

```python
#Try to think about how to create a column named `Rank_Level` like this:
#If the `world_rank` is between 1 and 100, then "First Tier Top University"
#If the `world_rank` is between 101 and 200, then "Second Tier Top University"
#If the `world_rank` is between 201 and 300, then "Third Tier Top University"
#Else："Other Top University"
```

```python
def world_rank_col(col_val):
    if col_val in range(1,101):
        val = 'First Tier Top University'
    elif col_val in range(101,201):
        val = 'Second Tier Top University'
    elif col_val in range(201,301):
        val = 'Third Tier Top University'
    else:
        val = 'Other Top University'
    return val
```

```python
df['rank_level'] = df['world_rank'].apply(world_rank_col)
```

```python
df.columns
```

```python
df = df[['rank_level','world_rank','institution','country','national_rank','score','year']]
```

```python
df
```

```python
##perfrom pivot table 
## there are 3 parameters 
## 1) values : the column that will be used in aggfunc
## 2) index : column that will be converted to index label 
## 3) columns : column that will be converted to columns header 

pvt_tb = df.pivot_table(values = 'score' , index = 'country' , columns = 'rank_level' , aggfunc = np.nanmean)
```
