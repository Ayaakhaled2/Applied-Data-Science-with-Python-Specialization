```python
def groupby_chars(col_name):
    if col_name[0] in ['A' , 'C' , 'F' , 'G' , 'H' , 'I']:
        val = 0
    elif col_name[0] in ['K' , 'L' , 'M' , 'N' , 'O' , 'P']:
        val = 1
    else:
        val = 2 
    return val
```
